## Problems based on Euler Equations

```python
from sympy import *
x = var('x')
y = Function('y')
dy = diff(y(x), x)

F = dy**2 + x * y(x)
print('The functional is ', F)
euler = diff(F, y(x)) - diff(diff(F, dy))
print('Differential Equation obtained is: ', euler)

sol = dsolve(euler, y(x), ics={y(0): 0, y(1): 0})
print('Solution is: ', sol)
```

## Isoperimetric Problems

```python
from sympy import *
x = var('x')
l = symbols('l')     #lagrange multiplier
y = Function('y')
dy = diff(y(x), x)

#To define the functional and the constraint
I = dy**2     # Functional
J = y(x)      # Constraint Function

G = I + l*J     # New functional
print('The new functional with the given constraint is ', G)

# Using the new functional in euler equation
euler = diff(G, y(x)) - diff(diff(G, dy), x)     
print('Substituting in euler equation: ', euler)

# Solution of the euler equation
sol1 = dsolve(euler, ics={y(0): 0, y(1): 1})     
# Solving the ODE with conditions
print('Solution of the Euler Equation subject to conditions is : ', sol1)

#To find the lagrange multiplier: 
#Above solution becomes the integrand of the constraint.
sol2 = Eq(integrate(sol1.rhs, (x, 0, 1)), 1)
print('Solution integral constraint is wrt the solution: ', sol2)

#To find the lagrange multiplier
lambd = solve(sol2, l)[0]
print('Lagrange multiplier is', lambd)

#Substitution of the lambda value in the solution
sol = sol1.subs(l, lambd)
print('The extremal is: ', sol)

#Plotting the extremal
plot(sol.rhs, (x, -10, 11))
```
