# Differental Equations - Solvable for p

1. Solve $x^2 p^2 + xyp - 6y^2 = 0$

```python
from sympy import *
x, p, C1, Y = symbols('x p C1 Y')
y = Function('y')
eq1 = x**2 * p**2 + x*y(x)*p - 6*y(x)**2
sol1 = solve(eq1, p)
p = diff(y(x), x)
sol2 = 1
for i in range(len(sol1)):
    deq = dsolve(Eq(p, sol1[i]), y(x), hint = 'default')
    sol2 = (deq.lhs - deq.rhs) * sol2
print(f"The required solutions is {sol2} = 0")

#To plot the Solution
sol3 = sol2.subs([(y(x), Y), (C1, 1)])
simplify(sol3)
plot_implicit(Eq(sol3, 0))
```

2. Solve $x p^2 + (y - x) p - y = 0$

```python
from sympy import *
x, p, C1, Y = symbols('x p C1 Y')
y = Function('y')
eq1 = x * p**2 + (y(x) - x)*p - y(x)
sol1 = solve(eq1, p)
p = diff(y(x), x)
sol2 = 1
for i in range(len(sol1)):
    deq = dsolve(Eq(p, sol1[i]), y(x), hint = 'default')
    sol2 = (deq.lhs - deq.rhs) * sol2
print(f"The solutions for the given Differential Equation is {sol2} = 0")

#To plot the Solution
sol3 = sol2.subs([(y(x), Y), (C1, 1)])
simplify(sol3)
plot_implicit(Eq(sol3, 0))
```

3. Solve $p(p + y) = x(x + y)$

```python
from sympy import *
x, p, C1, Y = symbols('x p C1 Y')
y = Function('y')
eq1 = p*(p + y(x)) - x*(x + y(x))
sol1 = solve(eq1, p)
p = diff(y(x), x)
sol2 = 1
for i in range(len(sol1)):
    deq = dsolve(Eq(p, sol1[i]), y(x), hint = 'default')
    sol2 = (deq.lhs - deq.rhs) * sol2
print(f"The solutions for the given Differential Equation is {sol2} = 0")

#To plot the Solution
sol3 = sol2.subs([(y(x), Y), (C1, 1)])
simplify(sol3)
plot_implicit(Eq(sol3, 0))
```
