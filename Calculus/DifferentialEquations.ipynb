# Solution for Ordinary Differential Equation of First Order

`dsolve(eq, f(x), hint)` : Solves ordinary differential equation `eq` for function `f(x)`, using method `hint`.

Here, we mention the type of method that needs to be used as `hint = '<solving method>'` and thereby mentioning that the given equation is of first order. These functions are intended for internal use by `dsolve()` and others.

Defining of equations using `Eq`: `Eq` is used to create symbolic equalities. Suppose we want to express $y = 2x + 1$, then we use the statement `Eq(y , 2*x + 1)` i.e., `Eq(LHS , RHS)`. Notice that the LHS and RHS are seperates with a ',' (comma) and not with an '=' or '==' symbol.

Note: For a homogeneous equation, one can just use `Eq(F(x, y, dy))` which represents $F(x, y, \dfrac{dy}{dx}) = 0$. But, in general, it is always better to mention the RHS value to be zero (0) in cases like these, `Eq(F(x, y, dy), 0)`. ie., `Eq(expr, 0)`.

One can also define the ODE without the function `Eq()`.

## Solving by Variable Separable Method

To solve ODE by separable method, we use `hint = 'separable'`.

1. Solve $\dfrac{d}{dx}f(x) = sec(x)^{2}$ by separating the variables

```python
from sympy import *
y = Function('y')
x = symbols('x')
display(Eq(y(x).diff(x), sec(x)**2))
classify_ode(Eq(y(x).diff(x), sec(x)**2))
```

```python
from sympy import *
y = Function('y')
x = symbols('x')
eqn = Eq(y(x).diff(x), sec(x)**2)
sol = simplify(dsolve(eqn, y(x), ics={y(0): 0}, hint='separable'))
print(f'Solution for the given ODE is y = {sol.rhs}')
plot(sol.rhs, (x, 0, 1))
```

We can also use `print(sol)` as print statement, which gives output as `Eq(function, solution)`. For the above example, the output will be `Eq(y(x), tan(x))`, which means $y(x) = tan(x)$.\newpage

2. Solve $\dfrac{d}{dx}y(x) = \dfrac{1}{x^{2} + 1}$

```python
from sympy import *
y = Function('y')
x = symbols('x')
eqn = Eq(y(x).diff(x), 1/(x**2 + 1))
sol = simplify(dsolve(eqn, y(x), ics={y(0): 0}, hint='separable', simplify=True))
print(f'Solution for the given ODE is y = {sol.rhs}')
plot(sol.rhs, (x, 0, 10))
```

## Linear ODE

To solve 1st order linear differential equations, `hint = '1st_linear'` is used in `dsolve` statement.

1. Solve $\dfrac{dy}{dx} = 2\,y\,tan(x) + sin(x)$, given $y(0) = 1$

```python
from sympy import *
y = Function('y')
x = symbols('x')
eqn = Eq(y(x).diff(x), 2*y(x)*tan(x) + sin(x))
sol = simplify(dsolve(eqn, y(x), ics={y(0): 1}, hint='1st_linear'))
print(f'Solution for the given ODE is y = {sol.rhs}')
plot(sol.rhs, (x, 0, 1))
```

\newpage

2. Solve $\dfrac{dy}{dx} = \dfrac{-y + 2\,x^{3}}{2\,x}$, given $y(0) = 1$

```python
from sympy import *
y = Function('y')
x = symbols('x')
eqn = Eq(y(x).diff(x), (-y(x) + 2*x**3)/(2*x))
sol = simplify(dsolve(eqn, y(x), ics={y(0): 1}, hint='default'))
print(f'Solution for the given ODE is y = {sol.rhs}')
plot(sol.rhs, (x, 0, 5))
```

\newpage

3. Solve $(1 + x^{2})\,\dfrac{dy}{dx} + 2\,x\,y = 4\,x^{2}$ given $y(0) = 0$

```python
from sympy import *
y = Function('y')
x = symbols('x')
eqn = Eq((1 + x**2)*y(x).diff(x) + 2 * x * y(x), 4 * x **2)
sol = simplify(dsolve(eqn, y(x), ics={y(0): 0}, hint='1st_linear')) 
print(f'Solution for the given ODE is y = {sol.rhs}')
plot(sol.rhs, (x, 1, 10))
```

## Bernoulli Differential Equations

To solve Bernoulli differential equations, `hint = 'Bernoulli'` is used.

1. Solve $x\,\dfrac{dy}{dx} + (1 + x)\,y = y^{2} \, x^{2}$, given $y(-1) = 1$

```python
from sympy import *
y = Function('y')
x = symbols('x')
eqn = Eq(x*y(x).diff(x) + (1 - x)*y(x), y(x)**2 * x**2)
sol = simplify(dsolve(eqn, y(x), ics={y(-1): 1}, hint = 'Bernoulli')) 
print(f'Solution for the given ODE is y = {sol.rhs}')
plot(sol.rhs, (x, 4, 15))
```

\newpage

2. Solve $x\,\dfrac{dy}{dx} + y = y^{2}\,log(x)$, given $y(1) = 1$

```python
from sympy import *
y = Function('y')
x = symbols('x')
eqn = Eq(x*y(x).diff(x) + y(x), y(x)**2 * log(x))
sol = simplify(dsolve(eqn, y(x), ics={y(1): 1}, hint = 'Bernoulli'))
print(f'Solution for the given ODE is y = {sol.rhs}')
plot(sol.rhs, (x, 1, 15))
```

\newpage

3. Solve $\dfrac{dy}{dx} + y\,tan(x) = y^{3}\,sec(x)$, given $y(\pi) = 3$

```python
from sympy import *
y = Function('y')
x = symbols('x')
eqn = Eq(y(x).diff(x) + y(x)*tan(x), y(x)**3 * sec(x))
sol = simplify(dsolve(eqn, y(x), ics={y(pi): 3}, hint = 'Bernoulli')) 
print(f'Solution for the given ODE is y = {sol.rhs}')
plot(sol.rhs, (x, pi, 2*pi))
```

## Exact Differential Equations

To solves 1st order exact ordinary differential equation, `hint = '1st_exact'` is used.

1. Solve $(3x + 2y + 1)\dfrac{dy}{dx} + 4x + 3y + 1$, given $y(1) = 0$

```python
from sympy import *
y = Function('y')
x = symbols('x')
eqn = Eq((3*x + 2*y(x) + 1) * y(x).diff(x) + (4*x + 3*y(x) + 1), 0)
sol = simplify(dsolve(eqn, y(x), ics={y(1): 0}, hint = '1st_exact')) 
print(f'Solution for the given ODE is y = {sol.rhs}')
plot(sol.rhs, (x, 1, 6))
```

\newpage

2. Solve $x + \dfrac{y^{3}}{x^{2}} = \dfrac{3y^{2}}{x}\,\dfrac{dy}{dx}$, given $y(1) = 1$

```python
from sympy import *
y = Function('y')
x = symbols('x')
eqn = (x + (y(x)**3 / x**2) - (3 * y(x)**2 / x) * y(x).diff(x))
sol = simplify(dsolve(eqn, y(x), ics={y(1): 1}, hint = '1st_exact')) 
print(f'Solution for the given ODE is y = {sol.rhs}')
plot(sol.rhs, (x, 3, 10))
```
