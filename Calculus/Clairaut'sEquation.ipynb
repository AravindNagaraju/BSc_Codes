1. Solve $y = xp + p^2$

```python
from sympy import *
x, y, p, C = symbols('x y p C')
eq1 = y - x*p - p**2
sol1 = eq1.subs(p, C)
print(f'The general solution is {sol1} = 0')
eq2 = diff(eq1, p)
p1 = solve(eq2, p)
sol2 = eq1.subs(p, p1[0])
print(f'The singular solution is {sol2} = 0')

#To plot
plot_implicit(sol2)
```

3. Solve $(y - px)(p - 1) = p$

```python
from sympy import *
x, y, p, C = symbols('x y p C')
eq1 = (y - p*x)*(p - 1) - p
sol1 = eq1.subs(p, C)
print(f'The general solution is {sol1} = 0')
eq2 = diff(eq1, p)
p1 = solve(eq2, p)
sol2 = eq1.subs(p, p1[0])
print(f'The singular solution is {sol2} = 0')

#To plot
plot_implicit(sol2)
```
